## Group By.

###Ejercicio 1:

1. ¿Cuántos préstamos hay por cada estado y cuál es la media de la cantidad de dinero prestado?
- where: se filtran los registros antes de agruparlos.
- having: se filtran los grupos tras obtenerlos.

```python
%%sql
SELECT status, count(*) as total, avg (amount)
FROM LOAN
WHERE duration >= 24
GROUP BY status
HAVING total > 25 
ORDER BY total DESC
--Tengo que tener en el select y en el group by la x. 
--Para filtrar grupos con HAVING. 
```
